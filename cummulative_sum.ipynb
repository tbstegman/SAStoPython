# SAS program to produce a cummulative sum by group

```
***********************************************
*   Create a small data set to demonstrate    *
*   the concept                               *
***********************************************;
data data1;
  input group $ x;
  datalines;
  A 1
  A 2
  B 3
  B 4
  B 5
  C 6
  C 7
  C 8
  C 9
  ;

***************************************************
*   Create a variable to store the cummulative    *
*   sum by group                                  *
***************************************************;
data data2;
	set data1;
		by group;
	if first.group then cummulative=x;
	else cummulative + x;
run;
```

# Python code to produce the same data set

````python
# Import packages/libraries
import numpy as np
import pandas as pd
````

````python
# create the data frame that was used in the SAS program above
data1 = (pd.DataFrame({'group':list('AABBBCCCC'),
                        'x':range(1,10)}))
data1
````

````python
# Create a copy of the data frame
data2 = data1.copy()

# Create a column to store the cummulative sum by group
data2['cummulative'] = data2.groupby('group').cumsum()

data2
````
